```python
from ultralytics import YOLO  
import cv2 

model_path = "./update_modell.onnx"

model = YOLO(model_path)
cap = cv2.VideoCapture(0)

# Display Window setup
cap.set(3,1024) 
cap.set(4,768)  

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break  

    results = model(frame)
    result = results[0]

    annotated_frame = result.plot() 
    cv2.imshow("Rock-Paper-Scissors", annotated_frame)
    
    # ASCII Code of Esc Button
    # Press Esc to quit
    if cv2.waitKey(1) == 27: 
        break

cap.release()
cv2.destroyAllWindows()
```
